RESP: O Program Counter (PC) é o resgistrador que armazena o endereço da próxima instrução a ser executada, logo é muito importante para controlar a sequência de processamento das intruções. Ele se encontra dentro do processador.

RESP: (1) ADD Op: 
- RI <- (PC): 1ª leitura na memória 
- Buscar o operando para somar com o valor de R0: 2ª leitura na memória

(2) SUB Op
- RI <- (PC): 1ª leitura na memória
- Buscar operando para subtrair com o valor do R0: 2ª leitura na memória
- Armazenar o resultado no endereço do operando: 1ª Escrita na memória

(3) ADD Op1 Op2
- RI <- (PC): 1ª leitura na memória
- Buscar operando 1 e 2: 2ª e 3ª leitura
- Armazenar o resultado da soma do OP1 + OP2 no endereço do OP1: 1ª escrita

(4) INCR 
- RI <- (PC): 1ª leitura na memória

(5) LDA Op 
- RI <- (PC): 1ª leitura na memória
- Buscar operando para jogar pro R0: 2ª Leitura na memória 

** Ao total foram 12 ciclos de memória, sendo 2 de escrita e 10 de leitura.

RESP: É o REM, pois ele tem o tamanho do endereço da memória. Então se fizermos todas as combinações de bits com o tamanho do REM conseguimos achar quantas células a memória tem.

```python
Image(filename='Q12.png')
```

```python
# RESP
numCelulas = 2**16 
# Como tem 64K = 2^16 células, são necessários 16 bits de endereço para a memória

numInstrucoes =  2**8
#Como tem 256 = 2^8 instruções, são necessários 8 bits no código de operação

#Cada instrução tem o mesmo tamanho da célula e deve conter o código de operação + operando ou endereço. 
# Então: o tamanho da célula vai ser os bits do C.OP + END
tamCelula = 8 + 16 

print('O tamanho do REM e do PC é o mesmo: 16 bits')
print(f'O tamanho do REG e do RDM é o mesmo: {tamCelula} bits')
print(f'O total de bits da memória vai ser {numCelulas} x {tamCelula} = {numCelulas*tamCelula} = 2^16 x 24 bits')
```
